```python
import yfinance as yf
import pandas as pd

# Get the list of S&P 500 tickers
sp500_tickers = pd.read_html(
    'https://en.wikipedia.org/wiki/List_of_S%26P_500_companies'
)[0]['Symbol'].tolist()

# Fetch historical data for each ticker
data = {}
for ticker in sp500_tickers:
    try:
        df = yf.download(ticker, start='2022-01-01', end='2022-12-31')
        if not df.empty:
            data[ticker] = df
    except Exception as e:
        print(f"Failed to download data for {ticker}: {str(e)}")

# Concatenate the data into a single DataFrame
combined_data = pd.concat(data.values(), keys=data.keys(), axis=1)

# Save data to CSV
combined_data.to_csv('sp500_data_with_tickers.csv')
```

```python
import pandas as pd
import yfinance as yf

# Get the list of S&P 500 tickers
sp500_tickers = pd.read_html(
    'https://en.wikipedia.org/wiki/List_of_S%26P_500_companies'
)[0]['Symbol'].tolist()

# Fetch historical data for each ticker
data = []
for ticker in sp500_tickers:
    try:
        df = yf.download(ticker, start='2022-12-01', end='2022-12-31')
        if not df.empty:
            df['Ticker'] = ticker  # Add a new column with the ticker name
            data.append(df)
    except Exception as e:
        print(f"Failed to download data for {ticker}: {str(e)}")

# Concatenate the data into a single DataFrame
combined_data = pd.concat(data)

# Save data to CSV
combined_data.to_csv('sp500_data_with_tickers.csv')
```
